```python
from datasets import load_dataset

dataset = load_dataset("wikiann", "en")  # For Amharic
print(dataset["train"][0])
```

```python
import pandas as pd
df=pd.DataFrame(dataset['train'])
df=df[['tokens','ner_tags']]

df
```

```python
ner_tag=dataset['train'].features['ner_tags'].feature.names
label2id={}
id2label={}
for idx,label in enumerate(ner_tag):
    label2id[label]=idx
for idx,label in enumerate(ner_tag):
    id2label[idx]=label
print(label2id)
print(id2label)
```

```python
from transformers import AutoTokenizer
model='xlm-roberta-base'
tokenizer=AutoTokenizer.from_pretrained(model)
tokenizer(dataset['train']['tokens'],is_split_into_words=True)
```

```python
labels = dataset['train']['ner_tags']
print(labels)
```

```python
def tokenize_label(tokenizer, label2id, id2label):
    def tokenize_and_align_labels(batch):
        tokenized_inputs = tokenizer(batch["tokens"], truncation=True, is_split_into_words=True)

        all_labels = batch["ner_tags"]
        aligned_labels = []

        for i, labels in enumerate(all_labels):
            word_ids = tokenized_inputs.word_ids(batch_index=i)
            previous_word_idx = None
            label_ids = []

            for word_idx in word_ids:
                if word_idx is None:
                    label_ids.append(-100)
                elif word_idx != previous_word_idx:
                    label_ids.append(labels[word_idx])
                else:
                    current_label = id2label[labels[previous_word_idx]]
                    if current_label.startswith("B-"):
                        new_label = "I-" + current_label[2:]
                        new_label_id = label2id.get(new_label, labels[word_idx])
                        label_ids.append(new_label_id)
                    else:
                        label_ids.append(labels[word_idx])
                previous_word_idx = word_idx

            aligned_labels.append(label_ids)

        tokenized_inputs["labels"] = aligned_labels
        return tokenized_inputs

    return tokenize_and_align_labels
```

```python
tokenized_dataset = dataset.map(
    tokenize_label(tokenizer, label2id, id2label),  # ← pass the function, not the result!
    batched=True,
    remove_columns=dataset["train"].column_names
)
```

```python
from transformers import DataCollatorForTokenClassification
data_collator=DataCollatorForTokenClassification(tokenizer=tokenizer)
```

```python
import numpy as np
# Move the import inside the function to ensure it's available
def compute_metrics(logits_and_label):
    from evaluate import load
    metric=load('seqeval')
    logits,label=logits_and_label
    predictions=np.argmax(logits,axis=-1)
    # Filter out predictions and labels at padding indices (-100)
    true_label = [[id2label[t] for t in l if t != -100] for l in label]
    true_pred = [[id2label[p] for (p, t) in zip(prediction, label) if t != -100] for prediction, label in zip(predictions, label)]

    result=metric.compute(predictions=true_pred,references=true_label)
    return{
        'precision':result['overall_precision'],
        'accuracy':result['overall_accuracy'],
        'recall':result['overall_recall'],
        'f1':result['overall_f1']
    }
```

```python
from transformers import TrainingArguments,AutoModelForTokenClassification
model_tranied=AutoModelForTokenClassification.from_pretrained(
    model,
    id2label=id2label,
    label2id=label2id,
)
traning_args=TrainingArguments(
    output_dir='yohannes_fine_tune_model',
 eval_strategy = "epoch",
    num_train_epochs=3,
    learning_rate=2e-5,
    per_device_eval_batch_size=16,
    per_device_train_batch_size=16,
    weight_decay=0.01
)
```

```python
from transformers import Trainer

trainer = Trainer(
  # Model to train
  model=model_tranied,


  # Training and validation datasets
  train_dataset=tokenized_dataset["train"],
  eval_dataset=tokenized_dataset["validation"],

  # Tokenizer
  tokenizer=tokenizer,

  # Custom metric function
  compute_metrics=compute_metrics,

  # Data collator
  data_collator=data_collator,
    args=traning_args
)
```

```python
trainer.train()
```

```python
trainer.save_model('yohannes_jesus_fine_model')
```

```python
from transformers import pipeline
ner=pipeline(
    'token-classification',
    model='yohannes_jesus_fine_model',
    aggregation_strategy='simple',
    device=0
)
```
